# Traemos datos de aeropuerto y lo transformo a un dato espacial

```python
import pandas as pd 
airport=pd.read_csv(os.path.join("data","ne-airports.csv"))
airport.head()
```

```python
airport.drop(index=0, inplace=True)
airport.reset_index(drop=True, inplace=True)
keep=['name','type','latitude_deg', 'longitude_deg','elevation_ft','region_name','municipality']
airport=airport.loc[:,keep]
airport
#infoairports.info()
```

```python
airport.info()
```

```python
a_numerico=['latitude_deg', 'longitude_deg','elevation_ft']
airport[a_numerico]=airport.loc[:,a_numerico].apply(lambda x:pd.to_numeric(x))

airport.info()
```

```python
hoja_1=niger.plot(color='white', edgecolor='black')
airport.plot.scatter(x = 'longitude_deg', y = 'latitude_deg',ax=hoja_1)
```

```python
airport_GDF=gpd.GeoDataFrame(data=airport.copy(),
                 geometry=gpd.points_from_xy(airport.longitude_deg,
                                             airport.latitude_deg),crs=4326)
airport_GDF=airport_GDF.to_crs(2931)  # ya proyectado
```

```python
hoja_2=niger.plot(color="pink",edgecolor="orange")
airport_GDF.plot(ax=hoja_2,markersize=5.5)
```

```python
airport_GDF['type'].value_counts()
```

```python
# Renombro el nombre de la columna type para evitar problemas de código
airport_GDF.rename(columns={'type':'kind'},inplace=True)
```

```python
from folium import LayerControl

m=airport_GDF[airport_GDF.kind=='small_airport'].explore(color="red",name="small",show=False)
m=airport_GDF[airport_GDF.kind=='medium_airport'].explore(m=m, color="blue",name="medium",show=False)
m=airport_GDF[airport_GDF.kind=='closed'].explore(m=m, color="yellow",name="closed",show=False)
LayerControl(collapsed=False).add_to(m)
m
```

```python
ComN_niger=niger_comuni.cx[:,centroidY:]
ComS_niger=niger_comuni.cx[:,:centroidY]                           
ComW_niger=niger_comuni.cx[:centroidX,:]
ComE_niger=niger_comuni.cx[centroidX:,:]
```

```python
ComNS_niger=ComN_niger.overlay(ComS_niger, how="intersection",keep_geom_type=True)
ComNS_niger.plot()
```

```python
ComN_niger.sjoin(ComS_niger, how="inner", predicate='contains')
```

```python
# Hago lo mismo, pero para este y oeste
ComWE_niger=ComW_niger.overlay(ComE_niger, how="intersection",keep_geom_type=True)
ComWE_niger.plot(edgecolor='white',linewidth=0.1)
```

# Unir geometrías

```python
ComNS_niger.overlay(ComWE_niger,how="union",keep_geom_type=True)
```

```python
pd.concat([ComNS_niger,ComWE_niger],ignore_index=True)
```

```python
ComNS_niger.dissolve().overlay(ComWE_niger.dissolve(), how="union",keep_geom_type=True).dissolve().plot()
```

```python
NewNiger=ComNS_niger.dissolve().overlay(ComWE_niger.dissolve(), how="union",keep_geom_type=True).dissolve()
NewNiger
```

```python
NewNiger['zone']='middles'
NewNiger=NewNiger.loc[:,['NOM_COM_1_1','zone','geometry']]   
NewNiger
```
