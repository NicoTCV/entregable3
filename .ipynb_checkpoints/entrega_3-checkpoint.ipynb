```python
import os

import geopandas as gpd
from  fiona import listlayers

nigermap='https://github.com/NicoTCV/simpleplot_/raw/main/maps/NigerWorld2931.gpkg'

listlayers(nigermap)
```

# Llamo al mapa Niger

```python
niger=gpd.read_file(nigermap,layer='country')
niger_city=gpd.read_file(nigermap,layer='cities')
niger_river=gpd.read_file(nigermap,layer='rivers')
```

```python
niger.crs.to_epsg()==niger_city.crs.to_epsg()==niger_river.crs.to_epsg()
```

```python
hoja=niger.plot(facecolor="pink")
niger_city.plot(marker='.', color='red', markersize=15,
                ax=hoja)
niger_river.plot(edgecolor='blue', linewidth=0.5,
                 ax=hoja)
```

```python
# contorno de Niger
niger.boundary.plot()
```
